```python
def read_directory_imagehub(dataset_path):
    # Percentage of images to be used for the validation set
    print("Loading from ", dataset_path)
    percentage_test = 20
    p = percentage_test/100
    # Populate the folders
    count=0
    for pathAndFilename in glob.iglob(os.path.join(dataset_path, "*.jpg")):  
        title, ext = os.path.splitext(os.path.basename(pathAndFilename))
        if random.random() <=p :
            os.system(f"cp {dataset_path}/{title}.jpg data/images/valid")
            os.system(f"cp {dataset_path}/YOLO/{title}.txt data/labels/valid")
        else:
            os.system(f"cp {dataset_path}/{title}.jpg data/images/train")
            os.system(f"cp {dataset_path}/YOLO/{title}.txt data/labels/train")
        count=count+1
    print('loaded ',count, 'images')
    return 
```
